```python
import random
import config
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep

# Setting up Firefox WebDriver
firefox_binary_path = 'C:/Program Files/Mozilla Firefox/firefox.exe' # Replace backslashes with forward slashes
options = webdriver.FirefoxOptions()
options.binary_location = firefox_binary_path

service = webdriver.firefox.service.Service('C:/Users/james/Downloads/geckodriver.exe') # Replace backslashes with forward slashes
driver = webdriver.Firefox(service=service, options=options)

# Opening LinkedIn URL and Logging in
driver.get('https://www.linkedin.com/login')

# Wait for the page to load completely
sleep(5)

# Find the username input field and enter the username
username_input = driver.find_element(By.NAME, 'session_key')
username_input.send_keys(config.LINKEDINEMAIL)

# Find the password input field and enter the password
password_input = driver.find_element(By.NAME, 'session_password')
password_input.send_keys(config.LINKEDINPWD)

login_button = driver.find_element(By.XPATH, "//button[@type='submit']")
login_button.click()
sleep(2)
# Navigating to a random page within the search results
page_number = random.randint(20, 40)
search_url = f'https://www.linkedin.com/search/results/people/?geoUrn=%5B%22105149562%22%5D&keywords=analyst%20OR%20data&network=%5B%22S%22%5D&page={page_number}'
driver.get(search_url)

# Set implicit wait time for 10 seconds
driver.implicitly_wait(10)

# Sending connection requests on all pages of search results
request_count = 0 # Initialize request counter

# Navigating to a random page within the search results
maxreq = random.randint(11, 35)

while request_count < maxreq: # Loop continues until the request count is less than the maximum number of requests
    connect_buttons = driver.find_elements(By.XPATH, "//button/span[text()='Connect']")
    if not connect_buttons: # Break out of loop if no more Connect buttons
        break

    for connect_button in connect_buttons:
        connect_button.click()
        sleep(random.randint(5, 10))
        send_now_button = driver.find_element(By.XPATH, "/html/body/div[3]/div/div/div[3]/button[2]/span")
        send_now_button.click()
        sleep(random.randint(5, 10))
        request_count += 1 # Increment request counter after each successful request
        print(f"Sent {request_count} connection requests out of {maxreq}.")

        if request_count >= maxreq: # Break out of loop if request count exceeds maximum number of requests
            break

    # Go to the next page of search results
    try:
        next_button = driver.find_element(By.XPATH, "/html/body/div[5]/div[3]/div[2]/div/div[1]/main/div/div/div[2]/div/div[2]/div/button[2]/span")
        next_button.click()
        sleep(random.randint(5, 10))
    except:
        break

# Closing the Browser window
driver.quit()

```
