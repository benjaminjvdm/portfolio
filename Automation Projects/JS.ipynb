```python
import random
import config
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# Setting up Firefox WebDriver
firefox_binary_path = 'C:/Program Files/Mozilla Firefox/firefox.exe' # Replace backslashes with forward slashes
options = webdriver.FirefoxOptions()
options.binary_location = firefox_binary_path

service = webdriver.firefox.service.Service('C:/Users/james/Downloads/geckodriver.exe') # Replace backslashes with forward slashes
driver = webdriver.Firefox(service=service, options=options)

# set the url and query string
url = 'https://www.google.com/search?q=site%3Ajobvite.com+%7C+site%3Abamboohr.com+%7C+site%3Agreenhouse.io+%7C+site%3Ajazzhr.com+%7C+site%3Aworkable.com+%7C+site%3Alever.co+Data+Analyst+OR+Business+Analyst+OR+Junior+Data+Scientist+OR+Data+Science+Analyst'

# open the search page
driver.get(url)
time.sleep(random.uniform(1.0, 3.0))

# Click on the "Tools" button
tools_button = driver.find_element(By.XPATH, '//*[@id="hdtb-tls"]')
tools_button.click()

# Wait for the "Any Time" dropdown menu to appear and then click it
any_time_dropdown = driver.find_element(By.XPATH, '/html/body/div[6]/div/div[4]/div/div[2]/div/span[1]/g-popup/div[1]/div/div/div')
time.sleep(random.uniform(1.0, 3.0))
any_time_dropdown.click()

# Wait for the options in the dropdown menu to appear and then click the "Past year" option
past_week_option = driver.find_element(By.XPATH, '/html/body/div[6]/div/div[6]/div/g-menu/g-menu-item[4]/div/a')
time.sleep(random.uniform(1.0, 3.0))
past_week_option.click()
time.sleep(10)

# create empty lists to store data
names = []
links = []

# iterate over all pages
for i in range(0,20):
    # parse the html content
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # find all the links
    page_links = soup.find_all('a')

    # append the link urls and names to the respective lists
    for link in page_links:
        href = link.get('href')
        if (href is not None) and ("http" in href) and ("google.com" not in href) and ("google.co.za" not in href):
            names.append(link.text)
            links.append(href)

    # Find the element that represents the "Next" button and click it
    try:
        next_button = driver.find_element(By.XPATH, '//a[@id="pnnext"]')
        next_button.click()

        # add a random delay before visiting the next page
        time.sleep(random.uniform(2.0, 5.0))

    except:
        print("Next button not found. Ending loop.")
        break

# close the webdriver
driver.close()

# create a pandas dataframe with the name and link columns
results = pd.DataFrame({'Name': names, 'Link': links})

results
```

```python
results.to_csv('clickup.csv', index=False)
```
